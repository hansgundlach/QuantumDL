```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.font_manager import FontProperties
```

## Base QEA Illustration

```python

# Generate data points
N = np.logspace(0, 40, 1000)  # Create points from 10^0 to 10^40

# Calculate complexity
classical = N  # O(N)
quantum = 1e9 * np.sqrt(N)  # O(10^9 * sqrt(N))

# Create the plot
plt.figure(figsize=(12, 8))

# Plot the lines
plt.plot(N, classical, label='Classical  O(N)', linewidth=5)
plt.plot(N, quantum, label='Quantum  O(10^9N)', linewidth=5)

# Set log scales
plt.xscale('log')
plt.yscale('log')

# Add labels and title
plt.xlabel('N (Problem size)', fontsize=30, fontweight='bold')
plt.ylabel('steps', fontsize=30, fontweight='bold')

# Set axis limits
# plt.xlim(1, 1e24)
# plt.ylim(1, 1e24)

# Add threshold point
N_threshold = 1e18
plt.plot([N_threshold], [N_threshold], 'ro', markersize=15)  # Larger red dot at intersection
plt.vlines(N_threshold, 1, N_threshold, 'r', linestyles='dashed', linewidth=3)  # Thicker dashed line

# Add N* label with larger and bolder font
plt.annotate('N* (Threshold problem size)', 
            xy=(N_threshold, N_threshold),
            xytext=(N_threshold*2, N_threshold/3),
            color='black',
            fontsize=35,
            fontweight='bold')

# Add "Classical Faster" and "Quantum Faster" labels
plt.annotate('Classical Faster', xy=(1e0, 1e25), color='black', fontsize=30, fontweight='bold')
plt.annotate('Quantum Faster', xy=(1e30, 1e25), color='black', fontsize=30, fontweight='bold')

# Add arrows to show regions
plt.annotate('', xy=(1e1, 1e23), xytext=(1e18, 1e23),
            arrowprops=dict(arrowstyle='<->', color='black', lw=3))
plt.annotate('', xy=(1e18, 1e23), xytext=(1e40, 1e23),
            arrowprops=dict(arrowstyle='<->', color='black', lw=3))

# Add legend
font_properties = FontProperties()
font_properties.set_size(25)
font_properties.set_weight('bold')
plt.legend(loc="lower right", fontsize=25, prop=font_properties)

# Add grid
plt.grid(True, which="both", ls="-", alpha=0.5)
# plt.title("Quantum Economic Problem Size For a Given Year", fontsize=20)
plt.xticks(fontsize=25, fontweight='bold')


yticks = [10.0**0, 10.0**10, 10.0**20, 10.0**30, 10.0**40]
yticklabels = [r"$\mathbf{10^0}$", r"$\mathbf{10^{10}}$", r"$\mathbf{10^{20}}$", r"$\mathbf{10^{30}}$", r"$\mathbf{10^{40}}$"]
plt.yticks(yticks, yticklabels, fontsize=25)



plt.yticks(fontsize=25, fontweight='bold')
plt.xlim(1, 1e40)
plt.tight_layout()
plt.title("Quantum Economic Problem Size For a Given Year", fontsize=25, fontweight='bold')
plt.savefig("Figures/qea_size_diagram.png", dpi=300, bbox_inches='tight')
```

# Descriptive Diagram For Quantum Deep Learning

```python


n_limit = 1e40
# Generate data points
N = np.logspace(0, np.log10(n_limit), 1000)  # Create points from 10^0 to 10^40
# Calculate complexity
classical = N # O(N)
quantum = 1e13 * N**0.5 # O(10^9 * sqrt(N))
n_threshold = 1e26# 1e4
N_threshold = n_threshold
step_threshold = 1e26

step_limit = 1e40
# Create the plot
# plt.figure(figsize=(12, 8))
plt.figure(figsize=(12, 10))

# Plot the lines
plt.plot(N, classical, label=r'Classical O($\mathbf{N}$)', linewidth=5)
plt.plot(N, quantum, label=r'Quantum O($\mathbf{10^{13} \sqrt{N}}$)', linewidth=5)

# Set log scales
plt.xscale('log')
plt.yscale('log')

# Add labels and title
plt.xlabel('N (Search Size)', fontsize=35, fontweight='bold')
plt.ylabel('Steps', fontsize=35, fontweight='bold')
plt.title("Quantum Economic Problem Size For a Given Year", fontsize=25, fontweight='bold', pad=20)

# Set axis limits
# plt.xlim(1, 1e24)
# plt.ylim(1, 1e24)

# Add threshold point
plt.plot([N_threshold], [step_threshold], 'ro', markersize=15)  # Larger red dot at intersection
plt.vlines(N_threshold, 1, step_threshold, 'r', linestyles='dashed', linewidth=3)  # Thicker dashed line

# Add N* label with larger and bolder font
plt.annotate('N* (Threshold problem size)', 
            xy=(N_threshold, step_threshold),
            xytext=(N_threshold*1.2, 1e-4*step_threshold/3),
            color='black',
            fontsize=30,
            fontweight='bold')

# Add "Classical Faster" and "Quantum Faster" labels
plt.annotate('Classical Faster', xy=(1e0, 15*step_threshold), color='black', fontsize=25, fontweight='bold')
plt.annotate('Quantum Faster', xy=(n_limit/1e6, 15*step_threshold), color='black', fontsize=25, fontweight='bold')

# Add arrows to show regions
plt.annotate('', xy=(1e0, step_threshold), xytext=(.85*N_threshold, step_threshold),
            arrowprops=dict(arrowstyle='<->', color='black', lw=3))
plt.annotate('', xy=(1.2*N_threshold, step_threshold), xytext=(n_limit, step_threshold),
            arrowprops=dict(arrowstyle='<->', color='black', lw=3))

# Add legend
font_properties = FontProperties()
font_properties.set_size(25)
font_properties.set_weight('bold')
plt.legend(loc="lower right", fontsize=25, prop=font_properties)

# Add grid only at major ticks
plt.grid(True, which="major", ls="-", alpha=0.7)
plt.xticks(fontsize=25, fontweight='bold')


xticks = [10.0**0, 10.0**10, 10.0**20, 10.0**30, 10.0**40]
xticklabels = [r"$\mathbf{10^0}$", r"$\mathbf{10^{10}}$", r"$\mathbf{10^{20}}$", r"$\mathbf{10^{30}}$", r"$\mathbf{10^{40}}$"]
plt.xticks(xticks, xticklabels, fontsize=25)



plt.yticks(fontsize=25, fontweight='bold')
plt.xlim(1, n_limit)
plt.ylim(1, step_limit)
plt.tight_layout()
plt.savefig("figures/qea_oneyear_illustrative.png", dpi=300, bbox_inches='tight')
```
